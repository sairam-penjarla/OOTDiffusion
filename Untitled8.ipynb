```python
import os
import shutil
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import shutil
from google.colab import files
```

```python
def move_files_and_folders(source_folder, destination_folder):
    # Ensure destination folder exists
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate over items in the source folder
    for item in os.listdir(source_folder):
        # Form the full path of the item
        source_path = os.path.join(source_folder, item)

        # Check if it's a file or directory
        if os.path.isfile(source_path):
            # Move the file to the destination folder
            shutil.move(source_path, destination_folder)
        elif os.path.isdir(source_path):
            # Move the directory to the destination folder
            shutil.move(source_path, os.path.join(destination_folder, item))

def get_output_path(folder_name):
    i = 0
    while os.path.exists(new_folder := f"{folder_name}{'-'+str(i) if i else ''}"):
        i += 1
    return new_folder

def display_images(image1_path, image2_path):
    # Read images
    image1 = cv2.imread(image1_path)
    image2 = cv2.imread(image2_path)

    # Resize images while maintaining aspect ratio
    width = 500
    height1 = int(image1.shape[0] * width / image1.shape[1])
    height2 = int(image2.shape[0] * width / image2.shape[1])
    image1_resized = cv2.resize(image1, (width, height1))
    image2_resized = cv2.resize(image2, (width, height2))

    combined_image = np.hstack((image1_resized, image2_resized))

    # Display the combined image
    cv2_imshow(combined_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def download_folder(folder_path):
    zip_file_path = f"{folder_path}.zip"
    shutil.make_archive(folder_path, 'zip', folder_path)
    files.download(zip_file_path)
```

```python
def change_path(file_path):
  # Read the content of the file
  with open(file_path, 'r') as file:
      lines = file.readlines()

  # Modify the lines to remove "../"
  modified_lines = []
  for line in lines:
    if '_PATH = "../checkpoints/' in line:
      line = line.replace('_PATH = "../checkpoints/', '_PATH = "checkpoints/')
    modified_lines.append(lines)

  # Write the modified content back to the file
  with open(file_path, 'w') as file:
      file.writelines(modified_lines)

change_path('ootd/inference_ootd.py')
change_path('ootd/inference_ootd_dc.py')
change_path('ootd/inference_ootd_hd.py')
```

```python
import sys
from PIL import Image
from pathlib import Path
from run.utils_ootd import get_mask_location

from preprocess.openpose.run_openpose import OpenPose
from preprocess.humanparsing.run_parsing import Parsing
from ootd.inference_ootd_hd import OOTDiffusionHD
from ootd.inference_ootd_dc import OOTDiffusionDC

# PROJECT_ROOT = Path(__file__).absolute().parents[1].absolute()
# sys.path.insert(0, str(PROJECT_ROOT))

def apply(model_type, category, cloth_path, model_path, gpu_id, image_scale, n_steps, n_samples, seed):
  openpose_model = OpenPose(gpu_id)
  parsing_model = Parsing(gpu_id)

  category_dict = ['upperbody', 'lowerbody', 'dress']
  category_dict_utils = ['upper_body', 'lower_body', 'dresses']

  if model_type == "hd":
      model = OOTDiffusionHD(gpu_id)
  elif model_type == "dc":
      model = OOTDiffusionDC(gpu_id)

  if model_type == 'hd' and category != 0:
      raise ValueError("model_type 'hd' requires category == 0 (upperbody)!")

  cloth_img = Image.open(cloth_path).resize((768, 1024))
  model_img = Image.open(model_path).resize((768, 1024))
  keypoints = openpose_model(model_img.resize((384, 512)))
  model_parse, _ = parsing_model(model_img.resize((384, 512)))
  output_folder_path = get_output_path("images_output/predict")
  os.makedirs(output_folder_path, exist_ok=True)

  mask, mask_gray = get_mask_location(model_type, category_dict_utils[category], model_parse, keypoints)
  mask.save(f'{output_folder_path}/stage-1.jpg')

  mask = mask.resize((768, 1024), Image.NEAREST)
  mask.save(f'{output_folder_path}/stage-2.jpg')

  mask_gray = mask_gray.resize((768, 1024), Image.NEAREST)
  mask_gray.save(f'{output_folder_path}/stage-3.jpg')

  masked_vton_img = Image.composite(mask_gray, model_img, mask)
  masked_vton_img.save(f'{output_folder_path}/stage-4.jpg')

  images = model(
      model_type=model_type,
      category=category_dict[category],
      image_garm=cloth_img,
      image_vton=masked_vton_img,
      mask=mask,
      image_ori=model_img,
      num_samples=n_samples,
      num_steps=n_steps,
      image_scale=image_scale,
      seed=seed,
  )

  image_idx = 0
  for image in images:
    output_file_path = f'{output_folder_path}/out_{model_type}_{image_idx}.png'
    image.save(output_file_path)
    image_idx += 1
  return output_file_path
```

```python
MODEL_TYPE = "hd" # "hd" or "dc"
CATEGORY = 0 # 0:upperbody; 1:lowerbody; 2:dress
CLOTH_PATH = "/content/run/examples/garment/06123_00.jpg"
MODEL_PATH = "/content/run/examples/model/01861_00.jpg"

OUTPUT_PATH = apply(
    model_type = MODEL_TYPE,
    category = CATEGORY,
    cloth_path = CLOTH_PATH,
    model_path = MODEL_PATH,
    gpu_id = 0,
    image_scale = 2.0,
    n_steps = 20,
    n_samples = 4,
    seed = -1
)

display_images(MODEL_PATH, OUTPUT_PATH)
```
